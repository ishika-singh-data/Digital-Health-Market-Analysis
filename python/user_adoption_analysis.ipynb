```python
import pandas as pd

# Create the user adoption data directly
data = {
    "User_ID": [1, 2, 3, 4, 5],
    "Age_Group": ["18-30", "18-30", "31-45", "31-45", "46-60"],
    "Platform": ["Practo", "Tata 1mg", "Practo", "PharmEasy", "mfine"],
    "Usage_Frequency": [3, 4, 2, 3, 1]
}

df = pd.DataFrame(data)

# Save to CSV inside Colab
df.to_csv("user_adoption.csv", index=False)

print("CSV file created successfully")
print(df)
```

```python
import os
os.listdir()
```

```python
import matplotlib.pyplot as plt

avg_usage = df.groupby("Age_Group")["Usage_Frequency"].mean()

avg_usage.plot(kind='bar')
plt.xlabel("Age Group")
plt.ylabel("Average Usage Frequency")
plt.title("Average Digital Health Usage by Age Group")
plt.tight_layout()
plt.savefig("python_avg_usage_by_agegroup.png", dpi=300)
plt.show()
```
